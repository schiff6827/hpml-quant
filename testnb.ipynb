```python
import torch
import numpy as np

print("=== CUDA / PyTorch ===")
print(f"PyTorch: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA version: {torch.version.cuda}")
    print(f"GPU: {torch.cuda.get_device_name(0)}")
    print(f"GPU memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
    # quick matmul test
    a = torch.randn(1000, 1000, device='cuda')
    b = torch.randn(1000, 1000, device='cuda')
    c = a @ b
    print(f"GPU matmul test: OK ({c.shape})")

print("\n=== RAPIDS ===")
import cudf, cuml, cugraph
print(f"cuDF: {cudf.__version__}")
print(f"cuML: {cuml.__version__}")
print(f"cuGraph: {cugraph.__version__}")
# quick cudf test
gdf = cudf.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
print(f"cuDF test: OK ({len(gdf)} rows)")

print("\n=== Key packages ===")
import sklearn, scipy, pandas, polars, numpy, numba, dask
for pkg in [sklearn, scipy, pandas, polars, numpy, numba, dask]:
    print(f"{pkg.__name__}: {pkg.__version__}")

print("\n=== NumPy/PyTorch interop ===")
arr = np.random.randn(100)
t = torch.from_numpy(arr).cuda()
back = t.cpu().numpy()
print(f"np->torch->cuda->np roundtrip: {'OK' if np.allclose(arr, back) else 'FAIL'}")

print("\n=== cuML + PyTorch CUDA sharing ===")
import cupy as cp
print(f"CuPy: {cp.__version__}")
x_cp = cp.random.randn(100)
x_torch = torch.as_tensor(x_cp, device='cuda')
print(f"CuPy->PyTorch zero-copy: {'OK' if x_torch.shape[0] == 100 else 'FAIL'}")
```
